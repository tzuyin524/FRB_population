The root mean square noise fluctuaiton in Jy

$\Delta S_{\mathrm{sys}}
= \frac{\beta T_{\mathrm{sys}}}{G \sqrt{n_p N_{\mathrm{add}} t_{\mathrm{samp}} \Delta f}}
= C \sigma$


$SNR=\frac{S}{\Delta S_{\mathrm{sys}}}$

```python
import numpy as np

def S( beta, T_sys, G, n_p, N_add, t_samp, delta_f):
    """
    Compute the signal-to-noise ratio (SNR).

    Parameters
    ----------
    S : detected flux
    beta : is a factor accounting for losses due to 1-bit digitization
    T_sys : system temperature ~30 K
    G : antenna gain ~0.7 K Jy^-1
    n_p : number of polarisations 2
    N_add : number of added samples
    t_samp : sampling time ~1ms
    delta_f : bandwidth 288 MHz
    """
# SNR = 10
    delta_S_sys = (beta * T_sys) / (G * np.sqrt(n_p * N_add * t_samp * delta_f)) #sigma
    return 10*delta_S_sys
```

```python
bandwidth = 340*10**6 # Hz
delta_t = 0.001 # s
Gain = 0.7 
Tsys = 25
flux = S(np.sqrt(np.pi/2),Tsys,Gain,2,1,delta_t,bandwidth) # parameters for prakes # flux density
print(flux,"Jy")
```

```python
L = 10**40 # erg s^-1
S = flux*10**(-23) #  
F = S*bandwidth #
d = np.sqrt(L/(4*np.pi*F)) # cm
dmax = d/(3.086*10**24) # Mpc
print (dmax,'Mpc')

```
